```python
df = df.fillna("")
```

```python
import collections
print(df.shape)
print(collections.Counter(df['description_material'].values.tolist())) # check labels all exist 
```

```python
# drop rows that have count less than 10

label_counts = df['description_material'].value_counts()

# Get labels where the frequency is less than 10
labels_to_filter = label_counts[label_counts < 10].index

# Filter out rows with labels that occur less than 10 times
df= df[~df['description_material'].isin(labels_to_filter)]
print(df.shape)
print(collections.Counter(df['description_material'].values.tolist()))
```

```python
# filter out too much frequent label rows
max_frequency = 15000

# Calculate the current frequency of values in the 'Category' column
value_counts = df['description_material'].value_counts()

# Identify values with frequency greater than max_frequency
values_to_limit = value_counts[value_counts > max_frequency].index

# Create an empty DataFrame to store the selected rows
selected_rows = pd.DataFrame(columns=df.columns)

# Randomly select up to 5000 rows for each value exceeding the limit
for value in values_to_limit:
    rows_with_value = df[df['description_material'] == value]
    random_sample = rows_with_value.sample(min(len(rows_with_value), max_frequency), random_state=42)
    selected_rows = selected_rows.append(random_sample)

# Select all rows where frequency is less than or equal to max_frequency
remaining_rows = df[df['description_material'].apply(lambda x: value_counts[x] <= max_frequency)]

# Concatenate the selected rows with the remaining rows
final_df = pd.concat([selected_rows, remaining_rows])

print(collections.Counter(final_df['description_material'].values.tolist())) # check labels all exist 
```

```python
final_df.shape
df = final_df
```

```python
from sklearn.model_selection import train_test_split

train_ratio = 0.7
train_df, dev_test_df = train_test_split(df, test_size=1 - train_ratio, random_state=42, stratify=df['description_material'])
dev_df, test_df = train_test_split(dev_test_df, test_size=0.5, stratify=dev_test_df['description_material'], random_state=42)
```

## Generate dataset for Text Entailment Task

```python
## 1. Use simple Name:Value 
import re
import random 
import collections
```

```python
# Define a function to concatenate values from specified columns
columns_to_concat = [x for x in df.columns if x != 'description_material' and x.startswith('description')]  # Exclude 'description_material'
url_pattern = r'https?://[\n\S]+\b'

def concatenate_cols(dataframe, verbose=False):
    # concatenate all the column fields 
    values = []
    for index, row in dataframe.iterrows():
        # Access row values by column name
        concatenated_text = ''
        for column in columns_to_concat:
            if str(row[column]) != '':
                if verbose is False:
                    concatenated_text += description_to_text[column] + ':' + str(row[column]) + ", "
                else:
                    concatenated_text += description_to_text[column] + ' is ' + str(row[column]) + ", "
        # Use the re.sub() function to replace URLs with an empty string
        cleaned_text = re.sub(url_pattern, '', concatenated_text)
        values.append(concatenated_text)
    return values
```

```python
# 1. generate concatenated text that does not have any description (non-language like)
#  [Col Name] : [Col Value] / Material : [Label]
```

```python
def generate_text_A(dataframe, train=True):
    dataframe['concatenated_text_A'] = concatenate_cols(dataframe, False)
    # Add hypothesis to text
    prefix = 'mat:'
    values = []
    neg_values = [] # generated negative examples
    gold_labels = list(set(dataframe['description_material'].values.tolist())) # labels that exist in this dataframe
    for index, row in dataframe.iterrows():
        # Add label as hypothesis (strip out the prefix and remove trailing comma)
        values.append('<s>' + row['concatenated_text_A'][:-2] + '</s><s>Material:' + str(row['description_material']) + '.</s>')
        # randomly select negative label from only seen labels 
        if train is True:
            negative_label = random.choice([label for label in gold_labels if label!= row['description_material'] and label not in row['label_list']])
            # generate negative example 
            neg_values.append('<s>' + row['concatenated_text_A'][:-2] + '</s><s>Material:' + negative_label + '.</s>')
    dataframe['concatenated_text_A'] = values
    if train is True:
        dataframe['negative_sample_A'] = neg_values
    return dataframe 
```

```python
# 2. Use more verbose version : [Col Name] gives [Col Value] / The material of this sample Material is [Label]
```

```python
def generate_text_B(dataframe, train=True):
    dataframe['concatenated_text_B'] = concatenate_cols(dataframe, True)
    # Add hypothesis to text
    prefix = 'mat:'
    values = []
    neg_values = []
    gold_labels = list(set(dataframe['description_material'].values.tolist())) # labels that exist in this dataframe
    for index, row in dataframe.iterrows():
        # Add label as hypothesis (strip out the prefix! )
        values.append('<s>' + row['concatenated_text_B'][:-2] + '</s><s>The material of this physical sample is ' + str(row['description_material'])+ '.</s>')
         # randomly select negative label
        if train is True:
            negative_label = random.choice([label for label in gold_labels if label!= row['description_material'] and label not in row['label_list']])
            # generate negative example
            neg_values.append('<s>' + row['concatenated_text_B'][:-2] + '</s><s>The material of this physical sample is ' + negative_label+ '.</s>')
    dataframe['concatenated_text_B'] = values
    if train is True:
        dataframe['negative_sample_B'] = neg_values
```
